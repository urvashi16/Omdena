```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
with open("allData.pkl","rb") as f:
    data=pickle.load(f)
```

```python
data.keys()
```

```python
#View the data
data['Resources']
```

```python
#storing files under Resources to res 
res = data['Resources']
```

```python
#Viewing the shape of data
res.shape
```

```python
#Describing the data
res.describe()
```

```python
res=res.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10',
       'Unnamed: 11'], axis=1)
```

```python
#Describing the data
res.describe()
```

```python
#Checking for number of unique data entries in each column
res.nunique()
```

```python
#Checking for unique data entries in each column
for col in res.columns:
    print('{} : {}'.format(col,res[col].unique()))
```

```python
#Checking the data type of each column
res.dtypes
```

```python
#Displaying all the column headings
res.columns
```

```python
#Renaming
res.rename(columns={'%Time ': 'Time'}, inplace=True)
res.head()
```

```python
#changing time to its standard format 
res['Time'] = pd.to_datetime(res['Time'], unit='D', origin='1899-12-30')
```

```python
#Checking for Null values in the data
res.isnull()
```

```python
#Checking for the sum of the null values in the data
res.isnull().sum()
```

```python
#Getting all the column headings in a list
list(res.columns.values)
```

```python
#Storing all the listed column headings excluding 'Time' in  a variable 
list_res=list(res.columns.values)
list_res.remove('Time')
```

```python
#Converting all the int data types to float 
for i in list_res:
    res[i] = res[i].astype(float)
```

```python
#Printing data types of each column
res.dtypes
```

```python
#Finding Correlation
correlation=res.corr()
correlation
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(res.corr(),cbar=True,annot=True,cmap='Oranges')
```
